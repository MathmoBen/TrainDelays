```python
reasons_csv = pd.read_excel('https://sacuksprodnrdigital0001.blob.core.windows.net/historic-delay-attribution/Reference%20Files/Transpareny%20Page%20Attribution%20Glossary.xlsx', sheet_name='Incident Reason')[['Incident Reason', 'Incident Reason Description']]
reasons_csv.rename(columns={'Incident Reason': 'INCIDENT_REASON', 'Incident Reason Description': 'INCIDENT_REASON_DESCRIPTION'}, inplace=True)
reasons_csv
```

```python
def fill_incident(incident, react):
    if pd.isnull(incident):
        return react
    return incident
    
df['INCIDENT_REASON'] = df.apply(lambda x: fill_incident(x['INCIDENT_REASON'], x['REACTIONARY_REASON_CODE']), axis=1)
df
```

```python
df.drop(columns='INCIDENT_REASON_DESCRIPTION', inplace = True)
```

```python
df = df.merge(reasons_csv, on='INCIDENT_REASON')
df
```

```python
retain_columns = ['EVENT_DATETIME', 'TRAIN_SERVICE_CODE_AFFECTED', 'ENGLISH_DAY_TYPE', 'APP_TIMETABLE_FLAG_AFF', 'TRAIN_SCHEDULE_TYPE_AFFECTED', 'TRACTION_TYPE_AFFECTED','UNIT_CLASS_AFFECTED', 'INCIDENT_REASON', 'Lat_OR', 'Lon_OR', 'Lat_DES', 'Lon_DES', 'PFPI_MINUTES' ]
```

```python
trim_df = df[retain_columns].copy()
trim_df.dropna(subset='PFPI_MINUTES', inplace=True) # drop those rows with NaN delays (why they are there in first place?!)
trim_df = trim_df[trim_df.PFPI_MINUTES != 0] # drop those with 0 minutes (same Question as above)
trim_df.loc[:, 'TRAIN_SERVICE_CODE_AFFECTED'] = trim_df.TRAIN_SERVICE_CODE_AFFECTED.astype(str) # converting to string not to confuse ourselves by scaling them accidentally
trim_df.loc[:, 'TRAIN_SCHEDULE_TYPE_AFFECTED'] = trim_df.TRAIN_SCHEDULE_TYPE_AFFECTED.apply(lambda x: 'STP' if 'STP' in x else 'LTP') # four categories here - LTP and the others (V)STP which I group them together. how they are short term planned is not our concern
trim_df = trim_df.dropna(subset= 'TRACTION_TYPE_AFFECTED') # dropping those without traction type (diesel or electronic - purely input mistake they must be either of each)
trim_df = trim_df[trim_df.TRACTION_TYPE_AFFECTED != 'L'] # wtf is L type :cry: (good there are only 3/4 rows)
trim_df.loc[:, 'UNIT_CLASS_AFFECTED'] = trim_df.UNIT_CLASS_AFFECTED.astype('Int64').astype(str) # same ideology for TSC

def is_DMS(traction_type, unit_class):
    if traction_type == 'DMS':
        return 172
    return unit_class
trim_df.loc[:, 'UNIT_CLASS_AFFECTED'] = trim_df.apply(lambda x: is_DMS(x['TRACTION_TYPE_AFFECTED'], x['UNIT_CLASS_AFFECTED']), axis=1) # mapping class 172 for Diesel locos as per described on wiki https://en.wikipedia.org/wiki/London_Overground#Past_fleet
trim_df = trim_df[trim_df.UNIT_CLASS_AFFECTED != '<NA>'] # drop those EMU locos with no specified unit class - may revisit later if domain knowledge available
trim_df.loc[:, 'INCIDENT_REASON'] = trim_df.INCIDENT_REASON.apply(lambda x: x[0]) # categorising incident code https://wiki.openraildata.com/index.php/Delay_Attribution_Guide
```

```python
# testing cyclical feature engineering for time-related columns

trim_df['MONTH_sin'] = np.sin(2*np.pi*(trim_df.EVENT_DATETIME.dt.month)/12)
trim_df['MONTH_cos'] = np.cos(2*np.pi*(trim_df.EVENT_DATETIME.dt.month)/12)
trim_df['DAY_sin'] = np.sin(2*np.pi*(trim_df.EVENT_DATETIME.dt.day)/31)
trim_df['DAY_cos'] = np.cos(2*np.pi*(trim_df.EVENT_DATETIME.dt.day)/31)
trim_df['HOUR_sin'] = np.sin(2*np.pi*(trim_df.EVENT_DATETIME.dt.hour)/24)
trim_df['HOUR_cos'] = np.cos(2*np.pi*(trim_df.EVENT_DATETIME.dt.hour)/24)
trim_df['MINUTE_sin'] = np.sin(2*np.pi*(trim_df.EVENT_DATETIME.dt.minute)/60)
trim_df['MINUTE_cos'] = np.cos(2*np.pi*(trim_df.EVENT_DATETIME.dt.minute)/60)
# all custom transformers
```
