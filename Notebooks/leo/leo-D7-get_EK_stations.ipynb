```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv('../data/EK_all.csv')
df
```

```python
station_or = df['STATION_OR'].unique()
station_des = df['STATION_DES'].unique()
```

```python
unique_station = pd.DataFrame(set(list(station_or) + list(station_des)))
unique_station.rename(columns={0: 'Station Name'}, inplace=True)
unique_station
```

```python
lookup = pd.read_csv('../mapping_data/stanox-lookup-lc.csv')
lookup
```

```python
unique_station = unique_station.merge(lookup, on='Station Name', how = 'inner').drop_duplicates(subset='Station Name').reset_index(drop=True)
```
